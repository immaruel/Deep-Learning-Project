# Pivot Table

상관분석을 위해 합친 파일을 matrix로 표현

```python
matrix = data.pivot_table(index = 'userId', columns='title', values='rating')

matrix.head(20)
```

# Pearson Correlation

https://namu.wiki/w/%EC%83%81%EA%B4%80%20%EA%B3%84%EC%88%98?from=%ED%94%BC%EC%96%B4%EC%8A%A8%20%EC%83%81%EA%B4%80%20%EA%B3%84%EC%88%98#s-2

일정 규칙에 따라 상관도 점수를 매기는 함수 작성 (세부규칙 확인 필요)

```python
'''
장르 및 캐스팅 종합
- 가중치는 조절 가능
'''

GENRE_WEIGHT = 0.1
CASTING_WEIGHT = 0.1

def pearsonR(s1, s2):
    s1_c = s1 - s1.mean()
    s2_c = s2 - s2.mean()
    return np.sum(s1_c * s2_c) / np.sqrt(np.sum(s1_c ** 2) * np.sum(s2_c ** 2))

def recommend(input_musical, matrix, n, similar_genre=True, similar_casting=True):
    input_genre = meta[meta['title'] == input_musical]['genre'].iloc(0)[0]
    input_casting_pre = meta[meta['title'] == input_musical]['casting'].iloc(0)[0]
    input_casting_list = input_casting_pre[1:-1].split(', ')
    input_casting = []
    for i in input_casting_list:
        i = i[1:-1]
        input_casting.append(i)

    result = []
    for title in matrix.columns:
        if title == input_musical:
            continue

        # rating comparison
        cor = pearsonR(matrix[input_musical], matrix[title])
        
        # genre comparison
        if similar_genre and len(input_genre) > 0:
            temp_genre = meta[meta['title'] == title]['genre'].iloc(0)[0]

            same_count = np.sum(np.isin(input_genre, temp_genre))
            cor += (GENRE_WEIGHT * same_count)
            
        # casting comparison
        if similar_casting and len(input_casting) > 0:
            temp_casting_pre = meta[meta['title'] == title]['casting'].iloc(0)[0]
            
            temp_casting_list = temp_casting_pre[1:-1].split(', ')
            temp_casting = []
            for j in temp_casting_list:
                j = j[1:-1]
                temp_casting.append(j)
            
            count = 0
            name_list = []
            for i in temp_casting:
                for j in input_casting:
                    if i == j:
                        name_list.append(i)
                        #print("일치 : ", i, j)
                        count += 1
                    else:
                        continue
            
            same_count = count
            cor += (CASTING_WEIGHT * same_count)
        
        if np.isnan(cor):
            continue
        else:
            result.append((title, '{:.2f}'.format(cor), temp_genre, temp_casting))
            
    result.sort(key=lambda r: r[1], reverse=True)

    return result[:n]
```

# Prediction

```python
'''
genre & casting 기반 추천
- 가장 좋아하는 뮤지컬 하나 입력하면 배우 이름과 비슷한 사람들 상관분석을 통해 뮤지컬 추천
'''
musical_name = input('가장 좋아하는 뮤지컬을 입력하세요 :') # 맨 오브 라만차
recommend_result = recommend(musical_name, matrix, 10, similar_genre=True, similar_casting=True)

pd.DataFrame(recommend_result, columns = ['Title', 'Correlation', 'Genre', 'Casting'])
```
